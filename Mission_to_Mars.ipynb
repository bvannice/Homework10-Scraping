```python
base_hemisphere_url = "https://astrogeoloty.usgs.gov"
marsHemisphere_url = "https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
browser.visit(marsHemisphere_url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

hemisphere_image_urls = []

hemispheres = soup.find_all('div', class_="item")
```

```python
for hemisphere in hemispheres:
    imageDictionary = {}
    title = hemisphere.find("h3").text
    nextImage = hemisphere.find("div", class_="description").a["href"]
    imageLocation = base_hemisphere_url + nextImage
    
    browser.visit(imageLocation)
    
    image_html = browser.html
    image_soup = BeautifulSoup(image_html, "html.parser")
    
    url3 = image_soup.find('ul').a["href"]
    
    imageDictionary["title"] = title
    imageDictionary["img_url"] = base_hemisphere_url + url3
    print(imageDictionary["img_url"])
    
    hemisphere_image_urls.append(imageDictionary)   
```
